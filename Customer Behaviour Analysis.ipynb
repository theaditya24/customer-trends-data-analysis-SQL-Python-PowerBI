## 4. Handling Missing Values & Changing Column Names

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()) )
```

**`Why use Median instead of Mean for Missing Value Imputation?`**

- **`Median is less affected by outliers`**, so it represents the central value more accurately in skewed data.
- **`Mean can be distorted by extreme values`**, leading to biased imputations and reducing model performance.

```python
df.isnull().sum()
```

```python
# Here you can see column names are not in right condition because they have some spaces between them.
# We will replace spaces with _ and make them into lower case. 
# It will make them more readable and easy to access
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df.columns
```

```python
# Also column = purchase_amount_(usd) this column-name has extra info 
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amonut'})
df.columns
```

## 5. Feature Engineering

```python
### Creating a Column age_group
bins = [0, 25, 40, 60, 100]
labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].value_counts()
```

```python
### Creating a column purchase_frequency_days. Here we have column = frequency_of_purchases
### which tells us about the frequency at which a user generally purchase something but the problem is that it is in object form.

purchase_mapping = {'Fortnightly':14, 'Weekly':7, 'Monthly':30, 'Quarterly':90, 'Bi-Weekly':14, 'Annually':365, 'Every 3 Months':90 } 
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(purchase_mapping)

df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

### Since all the **`values of both the columns are equal`**. Hence we **`do not need both columns for our further process, can drop one of them.`**

```python
df = df.drop('promo_code_used', axis = 1)
```
